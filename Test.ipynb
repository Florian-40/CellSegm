```python
import SimpleITK as sitk 
import numpy as np 
import matplotlib.pyplot as plt 
import os
from COp_Net.scripts.IterativeInf import Inference
from COp_Net.scripts.AllMetrics import Results

%matplotlib notebook
```

```python
Inference('Images/probamap.nii.gz', 'Outputs', max_iter=5)
```

```python
# Read images to evaluate the performances
pred = sitk.ReadImage('Outputs/CopNet_5_0000.nii.gz')  #shape (1,512,512)
labels = sitk.ReadImage('Images/ground_truth.nii.gz')
proba_map = sitk.ReadImage('Images/probamap.nii.gz')
hemo = sitk.ReadImage('Images/hemo.nii.gz')
vessel = sitk.ReadImage('Images/vessel.nii.gz')

# Convert them to numpy array
pred = sitk.GetArrayFromImage(pred)[0,:,:]
labels = sitk.GetArrayFromImage(labels)
proba_map = sitk.GetArrayFromImage(proba_map)
hemo = sitk.GetArrayFromImage(hemo)
vessel = sitk.GetArrayFromImage(vessel)

                        
    
# Evaluate the performance of the current 2D or 3D image 
res = Results(pred, labels, proba_map, 'Images/background_orig.nii.gz', hemo, vessel)
res.LabellingMetrics(error_percentage = 0.15, slice_visualization = 0)
res.SegmentationMetrics()
```
